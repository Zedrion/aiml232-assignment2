```python
import random
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

np.random.seed(232)

X, y = fetch_openml("Fashion-MNIST",
                    version=1,
                    as_frame=False,
                    return_X_y=True)

from sklearn import metrics

def train_classifier(clf, X_train, y_train, X_val, y_val, epochs=10):
    """Train a classifier with warm_start in 1-epoch chunks.
    clf: sklearn.neural_network.MLPClassifier with warm_start=True and max_iter=1
    X_train: training data features
    y_train: training data labels
    X_val: validation data features
    y_val: validation data labels
    epochs: number of epochs to train for (default: 10)
    Returns: train_loss, val_loss
    """

    train_loss = []
    val_loss = []
    for ep in range(epochs):
        clf.fit(X_train, y_train)
        train_loss.append(metrics.log_loss(y_train, clf.predict_proba(X_train)))
        val_loss.append(metrics.log_loss(y_val, clf.predict_proba(X_val)))
        print(
            f"Epoch {ep+1}/{epochs} | "
            f"Train Loss: {train_loss[-1]:.4f} | "
            f"Val Loss: {val_loss[-1]:.4f}"
        )

    return train_loss, val_loss
```

```python

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=10_000, random_state=232)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

# 4.2.

```python
clf_a = MLPClassifier(
    hidden_layer_sizes=(256,) * 10,
    activation='logistic',
    solver='sgd',
    learning_rate_init=1e-4,
    alpha=0.0,
    warm_start=True,
    max_iter=1,
    random_state=232
)

train_loss_a, val_loss_a = train_classifier(clf_a, X_train, y_train, X_val, y_val, epochs=10)

plt.plot(range(1, 11), train_loss_a, label='Train Loss')
plt.plot(range(1, 11), val_loss_a, label='Validation Loss')
plt.title('Model A: Sigmoid Activation')
plt.xlabel('Epoch')
plt.ylabel('Log Loss')
plt.legend()
plt.show()
```

```python
clf_b = MLPClassifier(
    hidden_layer_sizes=(256,) * 10,
    activation='tanh',
    solver='sgd',
    learning_rate_init=1.0,
    alpha=0.0,
    warm_start=True,
    max_iter=1,
    random_state=232
)

train_loss_b, val_loss_b = train_classifier(clf_b, X_train, y_train, X_val, y_val, epochs=10)

plt.plot(range(1, 11), train_loss_b, label='Train Loss')
plt.plot(range(1, 11), val_loss_b, label='Validation Loss')
plt.title('Model B: Tanh Activation')
plt.xlabel('Epoch')
plt.ylabel('Log Loss')
plt.legend()
plt.show()
```
