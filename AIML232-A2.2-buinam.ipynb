# 2.1.

For gamma distribution with shape parameter $\alpha=2$ and scale parameter $\beta=1.25$: $\newline$
$\mu = \alpha\beta = 2\cdot1.25 = 2.5 \newline$
$\sigma^2 = \alpha\beta^2 = 2\cdot1.25^2 = 3.125$

# 2.2.

```python
import random 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(232)

x = np.random.gamma(shape=2, scale=1.25, size=1000000)
plt.figure(figsize=(12,8))
plt.hist(x, bins=200, density=True) # The choice of number of bins is a visualization decision, not theoretical
plt.title("Histogram of chocolate chip weights (Gamma(α=2, β=1.25))")
plt.xlabel("Weight")
plt.ylabel("Density")
plt.grid(True, which="both", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2.3.

```python
import random 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(232)

bag_sizes = [3, 30, 100]
n_bags = 10_000
titles = ["1", "2", "3"]

fig, ax = plt.subplots(3, 1, figsize=(12,10), sharex=True)

for i in range(0,len(bag_sizes)):
    pop = np.random.gamma(2, 1.25, (n_bags, bag_sizes[i]))
    means = pop.mean(axis=1)
    ax[i].hist(means, bins=25, density=True)
    ax[i].set_title(f"Histogram of sample means (bag size = {bag_sizes[i]}), {n_bags} bags")
    ax[i].set_ylabel("Density")

ax[-1].set_xlabel("Mean Weight")

plt.tight_layout()
plt.show()


```
